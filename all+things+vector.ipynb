```python
#Poole, 1.1 Exercises

import matplotlib.pyplot as plt
import numpy as np


#1, Draw the following vectors in std. position, in R^2

def cartesian_plane():
    fig, ax = plt.subplots(figsize=(10,10))
    plt.xlim (-10,10)
    plt.ylim (-10,10)
    ax.set_xticks(np.arange(-10,12,2))
    ax.set_yticks(np.arange(-10,12,2))
    for spine in ['left', 'bottom']:
        ax.spines[spine].set_position('zero')
    for spine in ['right','top']:
        ax.spines[spine].set_color('none')
    ax.grid(axis=10)
    ax.set_title('Vectors & Vector ops')
    
    return fig, ax

#a = [3,0]
fig, ax = cartesian_plane()
ax.arrow(0,0,3,0, head_width = 0.5, head_length = 0.5, color = 'red')
plt.show()
#b = [3,2]
fig, ax = cartesian_plane()
ax.arrow(0,0,2,3,head_width = 0.5, head_length = 0.5, color = 'red')
plt.show()
#c = [-2,3]
fig, ax = cartesian_plane()
ax.arrow(0,0,-2,3,head_width = 0.5, head_length = 0.5, color = 'red')
plt.show()
#d
fig, ax = cartesian_plane()
ax.arrow(0,0,-3,-2,head_width = 0.5, head_length = 0.5, color = 'red')
plt.show()



```
